#### Codificacion

```python
X = df.drop("RainTomorrow", axis=1).copy()
y = df.RainTomorrow.copy()
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
X_train
```

```python
y_train.value_counts(normalize=True)
```

```python
y_test.value_counts(normalize=True)
```

```python
X_train.WindGustDir.unique()
```

```python
X_test.WindGustDir.unique()
```

```python
#### train
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse_output=False).fit(X_train[["WindGustDir","RainToday"]]) # SOLO SE ENTRENA CON TRAIN <-- IMPORTANTE!!
```

```python
encoded = ohe.transform((X_train[["WindGustDir","RainToday"]]))
```

```python
encoded_df = pd.DataFrame(columns=ohe.get_feature_names_out(), data = encoded, index=X_train.index)
```

```python
X_train = pd.concat([X_train, encoded_df], axis="columns")
```

```python
X_train = X_train.drop(["WindGustDir","RainToday"], axis=1)
```

```python
#### test
```

```python
encoded_test = ohe.transform((X_test[["WindGustDir","RainToday"]]))
```

```python
encoded_df_test = pd.DataFrame(columns=ohe.get_feature_names_out(), data = encoded_test, index=X_test.index)
```

```python
X_test = pd.concat([X_test, encoded_df_test], axis="columns")
```

```python
X_test = X_test.drop(["WindGustDir","RainToday"], axis=1)
```

##### entrenamiento

```python
### arbol de decision
```

```python
seed = 42
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(max_depth=3, random_state=seed)
```

```python
%%time
clf.fit(X_train, y_train)
```

```python
%%time
train_pred = clf.predict(X_train) # devuelve 0 o 1 dependiendo de la probabilidad: si p > 0,5 = 1, caso contrario 0
test_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
X_train.shape
```

```python
print("Train")
print(classification_report(y_train, train_pred))
print("----------------------------------------------------------")
print("Test")
print(classification_report(y_test, test_pred))
```

```python
### knn
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
clf = KNeighborsClassifier()
```

```python
%%time
clf.fit(X_train_scaled, y_train)
```

```python
%%time
train_pred = clf.predict(X_train_scaled)
test_pred = clf.predict(X_test_scaled)
```

```python
print("Train")
print(classification_report(y_train, train_pred))
print("----------------------------------------------------------")
print("Test")
print(classification_report(y_test, test_pred))
```

```python
#### random forest
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(n_estimators=20, min_samples_split=10, min_samples_leaf=50, random_state=seed)
```

```python
train_pred = clf.predict(X_train)
test_pred = clf.predict(X_test)
```

```python
print("Train")
print(classification_report(y_train, train_pred))
print("----------------------------------------------------------")
print("Test")
print(classification_report(y_test, test_pred))
```

```python
y_train_proba = clf.predict_proba(X_train)[:,1] # cambiar a 0 si quieren la otra prob
y_test_proba = clf.predict_proba(X_test)[:,1]
```

```python
y_train_proba
```

```python
plt.figure(figsize=(9,5))
sns.histplot(y_train_proba, bins=100, kde=True, label="Train", stat="density", alpha=0.7)
sns.histplot(y_test_proba, bins=100, kde=True, label="Test", stat="density", alpha=0.5)
plt.axvline(0.5, color="red", linestyle="--")
plt.title("aca va el titulo")
plt.xlabel("probabilidad")
plt.ylabel("density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
threshold = 0.79
```

```python
y_train_pred_custom = (y_train_proba >= threshold).astype(int)
y_test_pred_custom = (y_test_proba >= threshold).astype(int)
```

```python
y_train_pred_custom
```

```python
print("train")
print(classification_report(y_train, y_train_pred_custom))
print("test")
print(classification_report(y_test, y_test_pred_custom))
```
